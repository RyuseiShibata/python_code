```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
gesture_list = ['a','b','c','d','e','f']
anova_list = [0.62,0.74,0.60,0.64,0.76,0.86]
random_list = [0.58,0.76,0.62,0.68,0.76,0.84]

left = np.arange(len(anova_list))  # numpyで横軸を設定
width = 0.3

plt.bar(left, anova_list, width=width, label='anova')
plt.bar(left+width, random_list, width=width, label='random forest')

plt.xticks(left + width/2, gesture_list)

plt.xlabel("gesture")
plt.ylabel("gesture accuracy")
plt.grid(True)
plt.legend()
```
